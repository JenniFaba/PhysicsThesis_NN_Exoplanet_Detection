```python
import juliet
import starsim
import numpy as np
import os
import imageio
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import time
from joblib import Parallel, delayed
import multiprocessing
import math
from pathlib import Path
from astropy.io import fits
```

```python
def processInput(t,i,ss,tmax):

    ss.rotation_period = 24.47 
    ss.spot_T_contrast = 700.
    ss.convective_shift = 1
    ss.differential_rotation = -2.66
    ss.inclination = 7.25
    ss.temperature_photosphere = 5772.
    ss.mass = 1 
    ss.radius = 1 
    ss.logg = 5.61 

    
    overlap=True
    while overlap: 
        Nspots=1 
        ss.spot_map=np.zeros([Nspots,7])
        
        for j in range(Nspots):
            ss.spot_map[j][0]=0
            ss.spot_map[j][1]=100
            ss.spot_map[j][2]=90.
            ss.spot_map[j][3]=0
            ss.spot_map[j][4]=41+0.3*i

        overlap=starsim.nbspectra.check_spot_overlap(ss.spot_map,0) #true if spots are overlapping

    ss.compute_forward(observables=['rv','lc'],t=t)
    rv=ss.results['rv'] 
    bi=ss.results['bis']
    fw = ss.results['fwhm']
    co = ss.results['contrast']
    ph = ss.results['lc']
    ff_ph= ss.results['ff_ph']
    ff_sp = ss.results['ff_sp']
    ff_pl = ss.results['ff_pl']
    ff_fc = ss.results['ff_fc']
    ccf = ss.results['CCF']

    ff_tot=100.-ff_sp-ff_pl

    
    np.savetxt('STARSIM_output_CCFra_{}.txt'.format(i), np.array(ss.results['CCF']).T,fmt='%.10f')
    
    if i==0: file_out= open('STARSIM_parameters2.txt', 'w')
    if i>0: file_out= open('STARSIM_parameters2.txt', 'a')           
    
    file_out.write(str(i) + ' ' +str(ss.spot_map[0][3]) +' ' +str(ss.spot_map[0][4]) +  '\n')
    file_out.close()

    
start=time.time()
Nsims=50
t=np.arange(0,13.1,0.5)
ss=starsim.StarSim(conf_file_path='starsim.conf')

tmin=t.min()
tmax=t.max() 
num_cores = multiprocessing.cpu_count()

Parallel(n_jobs=num_cores)(delayed(processInput)(t,i,ss,tmax) for i in range(Nsims))
```
